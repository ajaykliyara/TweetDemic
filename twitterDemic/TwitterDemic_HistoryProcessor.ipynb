```python
def getMappingLocation(tweetCoOrd, authorCoOrd):
    if tweetCoOrd:
        return str(tweetCoOrd['coordinates'][0]) + ',' + str(tweetCoOrd['coordinates'][1])
    if authorCoOrd:
        return str(authorCoOrd['lng']) + ',' + str(authorCoOrd['lat'])
    
    return '-69.942848,37.176654' #atlantic ocean location
    

def getTweetAttr(tweet):
    tweet_attr = {
        'id' : tweet.id,
        'created_at_utc' : tweet.created_at,
        'tweet_text' : tweet.full_text,
        'tweet_url' : "www.twitter.com/" + str(tweet.author.screen_name) + "/status/" + str(tweet.id),
        'geo_coordinates' : tweet.coordinates,
        'fav_count_or_likes' : tweet.favorite_count,
        'retweet_count' : tweet.retweet_count,
        'author_location' :  tweet.author.location,
        'author_geo_coordinates' : getGeoCord(tweet.author.location),
        'author_followers_count' : tweet.author.followers_count,
        'author_screen_name' : tweet.author.screen_name,
        'author_fav_count' : tweet.author.favourites_count,
        'geo' : tweet.geo,
        'retweeted' : tweet.retweeted,
        'source url' : tweet.source_url,
        'source' : tweet.source,
    }
    
    tweet_attr['mapping_location'] = getMappingLocation(tweet_attr['geo_coordinates'], tweet_attr['author_geo_coordinates'])
    
    return tweet_attr
```

```python
import tweepy
import csv
import pandas as pd
import csv
from datetime import datetime
import time

#attributes pulled
header_fields = ['id','created_at_utc','tweet_text','tweet_url','geo_coordinates','fav_count_or_likes', 'retweet_count' ,
                'author_location','author_geo_coordinates','author_followers_count','author_screen_name','author_fav_count','geo','retweeted','source url','source','mapping_location']

file_name = 'wegmans/tweets_' + str(datetime.now()) +'.csv'; #saving file


#write tweets to file
with open(file_name, 'w') as output_file:
    dict_writer = csv.DictWriter(output_file, fieldnames = header_fields)
    dict_writer.writeheader()
    
    ctr=0
    #loop through pulled in tweets and write to file
    for tweet in tweepy.Cursor(api.search,q="#wegmans", count=100, lang="en", tweet_mode='extended', include_entities = True).items(maxTweets):
        time.sleep(1)
        ctr = ctr + 1
        dict_writer.writerows([getTweetAttr(tweet)])


print('row written to file = {0}'.format(ctr)) # rows written to file
wr_file = pd.read_csv(file_name)               # read file back into pandas dataframe

print('file shape = ', wr_file.shape)
wr_file.head()
```
